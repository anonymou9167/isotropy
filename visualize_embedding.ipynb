```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
def plot_pca_layers(speed, noise, dd):
    fig = plt.figure(figsize=(18, 3))

    for i in range(6):
        data = np.load(f"result_embedding/{dd}_full_gpt2_layer{i}_speed{speed}_noise{5}.npy")
        y = np.vstack(data)
        pca = PCA(n_components=3).fit(y)
        y = pca.transform(y)
        var_ratio = sum(pca.explained_variance_ratio_)
        ax = fig.add_subplot(1, 6, i + 1, projection='3d')
        ax.scatter(y[:, 0], y[:, 1], y[:, 2], s=0.1, alpha=0.3, marker='.')
        ax.set_title(f'Layer {i} Var Ratio {var_ratio:.2f}')

    # plt.suptitle(f'PCA on LLM Layers {dd} Speed {speed} Noise {noise}')
    plt.tight_layout()
    plt.show()
```

```python
plot_pca_layers(speed=50, noise=18, dd='tdd')
```

```python
plot_pca_layers(speed=50, noise=5, dd='fdd')
```
