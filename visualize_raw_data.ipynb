```python
import hdf5storage
import matplotlib.pyplot as plt
import numpy as np
```

```python
prev_path = "./dataset/test_data/H_U_his_test.mat"     # path of dataset [H_U_his_test]
pred_path = "./dataset/test_data/H_U_pre_test.mat"      # path of dataset [H_U_pre_test]
pred_path_fdd = "./dataset/test_data/H_D_pre_test.mat"  # path of dataset [H_D_pre_test]
test_data_prev_base = hdf5storage.loadmat(prev_path)['H_U_his_test']
test_data_pred_base = hdf5storage.loadmat(pred_path)['H_U_pre_test']
test_data_pred_base_fdd = hdf5storage.loadmat(pred_path_fdd)['H_D_pre_test']
```

```python
test_data_prev_base.shape
```

```python
test_data_prev_10 = test_data_prev_base[[0], ...]
test_data_prev_100 = test_data_prev_base[[9], ...]

test_data_pred_10 = test_data_pred_base[[0], ...]
test_data_pred_100 = test_data_pred_base[[9], ...]

test_data_pred_10_fdd = test_data_pred_base_fdd[[0], ...]
test_data_pred_100_fdd = test_data_pred_base_fdd[[9], ...]
```

```python
s1 = test_data_prev_10[0, 0, :, 0, 0, 0,0]
s2 = test_data_prev_100[0, 0, :, 0, 0, 0,0]

s_in = [s1, s2]
s_out = [test_data_pred_10[0, 0, :, 0, 0, 0,0], test_data_pred_100[0, 0, :, 0, 0, 0,0]]
s_out_fdd = [test_data_pred_10_fdd[0, 0, :, 0, 0, 0,0], test_data_pred_100_fdd[0, 0, :, 0, 0, 0,0]]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(13, 5))

for i, s in enumerate(s_out):
    axs[0].plot(s, linewidth=2, marker='d')
axs[0].legend(['10 km/hour', '100 km/hour'], fontsize=18)

for i, s in enumerate(s_out_fdd):
    axs[1].plot(s, linewidth=2, marker='o')
axs[1].legend(['10 km/hour', '100 km/hour'], fontsize=18)

axs[0].set_xlabel('Time instance', fontsize=24)
axs[0].set_ylabel('Signal', fontsize=24)
axs[0].set_title('Dataset 1', fontsize=24)
axs[0].tick_params(axis='both', which='major', labelsize=18)

axs[1].set_xlabel('Time instance', fontsize=24)
axs[1].set_title('Dataset 2', fontsize=24)
axs[1].tick_params(axis='both', which='major', labelsize=18)

axs[0].grid()
axs[1].grid()

plt.show()
```
